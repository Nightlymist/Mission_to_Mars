```python
#Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Automated browser
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
# Visit the Mars NASA news site
url = 'https://data-class-mars-challenge.s3.amazonaws.com/Mars/index.html'
browser.visit(url)
```

```python
# Scrape the HTML table
html = browser.html
soup_mars = soup(html, 'html.parser')
mars_tb = soup_mars.find('table', class_='table')
print(mars_tb)
```

```python
# Scrape the HTML content inside table
mars_body = soup_mars.find('tbody')
for mars in mars_body:
    print(mars.text)
```

```python
# Creating list for data to be stored
mars_list = []
```

```python
# Scraping data through for loop to get each column
def get_summary():
    #Using global variable
    global mars_list
    #Obtaining html
    html = browser.html
    news_soup= soup(html, 'html.parser')
    
    #Selecting all table rows
    mars_data = soup_mars.find_all('tr', class_='data-row')
    
    #Creating list of dictionary based on table content
    for data in mars_data:
        row = data.find_all('td')
        
        #Creating dictionary format
        summary_dict = {
            'id': row[0].text,
            'terrestrial_date': row[1].text,
            'sol': row[2].text,
            'Is': row[3].text,
            'month': row[4].text,
            'min_temp': row[5].text,
            'pressure': row[6].text
            
        }
        mars_list.append(summary_dict)
get_summary()
```

```python
#Print out mars list
mars_list
```

```python
# Changing mars_list into dictionary
df = pd.DataFrame(mars_list)
df
```

```python
#Determining nulls are present
df.info()
```

```python
# Casting data types
df = df.astype({'id': 'int32','Is':'int32','terrestrial_date':'string','sol': 'int32','month': 'int32', 'min_temp': 'float32', 'pressure': 'float32'})
```

```python
# Formating date for terrestrial date
df.terrestrial_date = pd.to_datetime(df.terrestrial_date, format='%Y-%m-%d')
```

```python
#Checking for errors
df.info()
```

```python
# Determine the number of months exist on mars
df.month.max()
```

```python
# Dropping duplicates if present
df.drop_duplicates()
```

```python
#Counting how many days worth of data exist
df.sol.count()
```

```python
#Finding the minimum average for minimum temperature column by grouping each month
df.groupby(by=["month"]).mean().min()
```

```python
#Finding the maximum average for minimum temperature column by grouping each month
df.groupby(by=["month"]).mean().max()
```

```python
#Displaying minimum temperature per Martian month
months = df.groupby(by=["month"]).mean()
temp = months.loc[:, ["min_temp"]]
temp
# The 8th month is the maximum temperature
# The 3rd month is the minimum temperature
```

```python
#Displaying minimum temperature versus Martian Month through bar graph
plt.bar(temp.index, temp.min_temp)
plt.xlabel('Month')
plt.ylabel('Minimum Temperature(°C)')
plt.title('Month vs. Minimum Temperature')
plt.xticks(temp.index)
plt.gca().invert_yaxis()
```

```python
#Displaying pressure per Martian month
pressure = months.loc[:, ["pressure"]]
pressure
# The 9th month is maximum pressure
# The 6th month is minimum pressure
```

```python
# Determining the maximum pressure 
months.loc[:, ["pressure"]].max()
```

```python
# Determining the minimum pressure of mars for which month
months.loc[:, ["pressure"]].min()
```

```python
# Displaying the bar graph of pressure over the months of mars
plt.bar(pressure.index, pressure.pressure)
plt.xticks(pressure.index)
plt.xlabel('Month')
plt.ylabel('Pressure')
plt.title('Month vs. Pressure')
```

```python
#Visualizing the cycles of mars temperature each year
plt.xticks(range(0, 2000, 200))
plt.xlim(0, 2000)
plt.grid()
plt.plot(df.sol, df.min_temp)
# Visually the mars year is about 650 days from first peak to second peak
```
